```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing import text, sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import Conv1D, GlobalMaxPooling1D, MaxPooling1D
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
```

# Load data

```python

tcc_df = pd.read_csv("final_wrangle.csv")
print(tcc_df.head(10))

train_tcc, test_tcc = train_test_split(tcc_df, test_size=0.2)
```

```python
print(train_tcc.loc[train_tcc['toxic_classifier']==1].head(8))
comment_text = train_tcc['comment_text'].values
toxic_classifier = train_tcc['toxic_classifier'].values
```

# Tokenizing comment text data

```python
"""Here I tokenized my data so that I'd get the top 18_000 most frequently 
occuring words"""

max_input = 18000
max_comment_length = 350

x_tok = text.Tokenizer(max_input)
x_tok.fit_on_texts(list(comment_text))
x_tokenized = x_tok.texts_to_sequences(comment_text)
train_tcc_val = sequence.pad_sequences(x_tokenized, maxlen=max_comment_length)
```

```python
# Using GloVe for pre-trained embeddings -> https://downloads.cs.stanford.edu/nlp/data/glove.6B.zip
"""GloVe focuses on words co-occurrences over the whole corpus. 
Its embeddings relate to the probabilities that two words appear together."""

embedding_dims = 100
embeddings_index = dict()
file = open('glove.6B.100d.txt', encoding="utf8")
for line in file:
    
    vals = line.split()
    init_text = vals[0]
    coefficients = np.asarray(vals[1:], dtype='float32')
    embeddings_index[init_text] = coefficients
file.close()
print(f"We found {len(embeddings_index)} words with embeddings")
```

# Creating the embedding layer

```python
# In this section, I have created an embedding layer. The input constant is a matrix that I created first from the embedded vectors

embed_matrix = np.zeros((max_input, embedding_dims))
for text, index in x_tok.word_index.items():
    if index > max_input-1:
        break
    else:
        embed_vec = embeddings_index.get(text)
        if embed_vec is not None:
            embed_matrix[index] = embed_vec
```

```python
model = Sequential()
model.add(Embedding(max_input, embedding_dims, 
                    embeddings_initializer=tf.keras.initializers.Constant(embed_matrix),
                   trainable=False))

model.add(Dropout(0.2)) 
```

# Building the Conv1D model

```python
# The hyperparameters for the ID-convolutional model
filters = 250
filter_size = 3
hidden_layer_dims = 250

model.add(Conv1D(filters, filter_size, padding='valid'))
model.add(MaxPooling1D())
model.add(Conv1D(filters, 5, padding='valid', activation='relu'))
model.add(Dropout(0.2))
model.add(GlobalMaxPooling1D())
model.add(Dense(hidden_layer_dims, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))
model.summary()


```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

# Model training

```python
# Model training. I have further divided my training dataset to have an evaluation dataset 

x_train, x_eval, y_train, y_eval = train_test_split(train_tcc_val, toxic_classifier, 
                                                    test_size=0.1, random_state=1)
batch_size = 32

model.fit(x_train, y_train, batch_size=batch_size, 
          epochs=5, validation_data=(x_eval, y_eval))
```

# Evaluating the model

```python
# Model evaluation. On the last cell, the goal is have an actual visualisation of the predictions

x_test = test_tcc['comment_text'].values
x_test_tok = x_tok.texts_to_sequences(x_test)
x_testing = sequence.pad_sequences(x_test_tok, maxlen=max_comment_length)

y_testing = model.predict(x_testing, verbose=1, batch_size=32)
```
